### Data Visualisation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv(r"C:\Users\hilda\Downloads\Ice Cream Ratings.csv")
df1 = df1.set_index('Date')
df1
```

```python
#Can change different styles using the matplotlib
print(plt.style.available)
plt.style.use('seaborn-v0_8-pastel')
```

```python
# Without specification, plot will give you a line graph with all the columns
df1.plot()
```

```python
# With specification
df1.plot(kind = 'line')
```

```python
#Subplots will break the plots
df1.plot(kind = 'line', subplots = True)
```

```python
#Adding Titles
df1.plot(kind = 'line', title = 'Ice Cream Ratings')
```

```python
#Adding axes lqbels
df1.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
# Bar chata
df1.plot(kind = 'bar')
```

```python
# Stacked Bar chata
df1.plot(kind = 'bar', stacked = True)
```

```python
# You can determine the colums you want
df1['Flavor Rating'].plot(kind = 'bar')
```

```python
# Creating a horizontal bar chat
df1.plot.barh(stacked = True)
```

```python
#Scatterplots, s will enlarge size of dots and c will determine the colors of the ddots.
df1.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 500, c = 'red')
```

```python
#Histograms with bins representing the distribution
df1.plot.hist(bins = 20)
```

```python
#Boxplots
df1.boxplot()
```

```python
#areaplots and fig sie changes the size of the chart
df1.plot.area(figsize = (10,5))
```

```python
#Pie charts
df1.plot.pie(y = 'Flavor Rating', figsize = (10,5))

```
